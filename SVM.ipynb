```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in data
data = pd.read_csv('dataset.csv', sep=',', index_col=0)
data.head()
```

```python
data.reset_index(inplace=True)
pre = 0;
for i in range(len(data.index)):
    if(data['Close Price'][i] > pre):
        data.loc[i, 'target'] = 1;
    else:
        data.loc[i, 'target'] = 0;
    pre = data['Close Price'][i];
```

```python
data.set_index("Date" , inplace=True)

data.head(10)
```

```python
pred_data = data[['Open Price','Close Price','High Price','Low Price','target']]
pred_data.head()
```

```python
#get training set
train = pred_data[:'29-Dec-2017']
train.reset_index(inplace=True)
train = train.drop(['Date'], axis=1)

train.head()
```

```python
#get test set
test = pred_data['02-Jan-2018':]
test.reset_index(inplace=True)
test = test.drop(['Date'], axis=1)

test.head()
```

```python
y_train = train['target']

y_train.head()
```

```python
y_test = test['target']

y_test.head()
```

```python
x_train = train.drop(['target'], axis=1)

x_train.head()
```

```python
x_test = test.drop(['target'], axis=1)

x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)

X_train_std = scaler.transform(x_train)
X_test_std = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Custom kernel function
def my_kernel(X, Y):
    return np.dot(X, Y.T)
  
penalty = 0.05

svm = SVC(C=penalty, kernel="linear", probability=True)
svm.fit(x_train, y_train.values)
  
acc_rate = accuracy_score(y_test.values, svm.predict(x_test)) * 100
print("Penalty = %.2f, Accuracy = %.2f %%" % (penalty, acc_rate))
 
```
